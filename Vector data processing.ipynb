```python
from shapely.geometry import Polygon, LineString, Point
```

```python
 point1 = Point(47.783, 13.0718) # creating point 1
```

```python
point2 = Point (47.895, 12.949) # creating point 2
```

```python
line1 = LineString([p1,p2]) # creating line 1
```

```python
polygon1 = Polygon ([(47.895, 12.949), (50.895, 14.949), (45.895, 10.200) ]) # creating polygon 1
```

```python
polygon1
```

```python
geoseries = gpd.GeoSeries([point1,point2,line1,polygon1])
```

```python
geoseries
```

```python
df = gpd.GeoDataFrame(geoseries)
df
```

```python
df.columns = ['geometry'] #chaning the name of the column
df
```

```python
df.crs = 'EPSG:4326'
df
```
